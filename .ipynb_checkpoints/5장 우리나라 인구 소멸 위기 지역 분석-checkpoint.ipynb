```python
import pandas as pd
import numpy as np

import platform
import matplotlib.pyplot as plt

%matplotlib inline

path = "c:/Windows/Fonts/malgun.ttf"
from matplotlib import font_manager, rc
if platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font',family=font_name)
    
plt.rcParams['axes.unicode_minus'] = False
```

```python
population = pd.read_excel('./data/05. population_raw_data.xlsx',header=1)

# 비어있는 곳은 NaN 처리하지 않음
population.fillna(method='pad',inplace=True)

# 컬럼의 이름을 '광역시도','시도','인구수'로 수정
population.rename(columns = {'행정구역(동읍면)별(1)':'광역시도',
                            '행정구역(동읍면)별(2)':'시도',
                            '계':'인구수'},inplace=True)
# 소계 컬럼 삭제
population = population[(population['시도'] != '소계')]

population
```

```python
# 원본을 복사를 할지 안할지 여부
population.is_copy = False
# '항목'칼럼을 '구분'으로 이름 바꾸고 실제 데이터에도 적용
population.rename(columns = {'항목':'구분'},inplace=True)

# '구분' 칼럼에 있는 이름들을 바꿔준다.
# loc[바꿀칼럼 == '바꾸는 이름', 컬럼] = '바꿀 이름'
population.loc[population['구분']=='총인구수 (명)','구분'] = '합계'
population.loc[population['구분']=='남자인구수 (명)','구분'] = '남자'
population.loc[population['구분']=='여자인구수 (명)','구분'] = '여자'

population
```

## 5-3 인구소멸 위기 지역 계산하고 데이터 정리하기

```python
# 20-39세까지 있는 값들을 더해준다
population['20 - 39세'] = population['20 - 24세'] + population['25 - 29세'] + \
                            population['30 - 34세'] + population['35 - 39세']
# 65세 이상 있는 값들을 더해준다
population['65세이상'] = population['65 - 69세'] + population['70 - 74세'] + \
                        population['75 - 79세'] + population['80 - 84세']+ \
                        population['85 - 89세'] + population['90 - 94세'] + \
                        population['95 - 99세'] + population['100+']
population.head(10)
```

```python
# index는 '광역시도','시도'로 하고, 칼럼은 '구분', 안에 들어가는 값들은 '인구수','20-39세','65세 이상'이 들어간다.
pop = pd.pivot_table(population, index = ['광역시도','시도'], 
                     columns = ['구분'],
                     values = ['인구수','20 - 39세','65세이상'])
pop
```

```python
# 인구 소멸 비율 : 20-30대 여자 / (65세이상/2), 소멸비율은 파생변수
# 소명위기지역 : 소멸비율 < 1.0

pop['소멸비율'] = pop['20 - 39세','여자'] / (pop['65세이상','합계'] / 2)
pop.head()
```

```python
# '소멸위기지역'칼럼을 새로 만들고, '소멸비율'칼럼의 값이 <1이면 T/F가 뜨게끔 한다.
pop['소멸위기지역'] = pop['소멸비율'] < 1.0
pop.head()
```

```python
#'소멸위기지역' 칼럼의 값이 TRUE이면 그에 맞는 '시도'칼럼이 출력
# multiful columns에서 값 뽑기 : df.index_get_level_values(1)을 이용하여 소멸위기지역을 뽑아낼 수 있다
pop[pop['소멸위기지역']==True].index.get_level_values(1)
```

```python
# pivot_table에 의해 다단으로 구성된 인덱스를 다시 초기화
pop.reset_index(inplace=True)
pop.head()
```

```python
# '20-39세' .. 등 합쳐있던 컬럼을 나누어 준다
tmp_columns = [pop.columns.get_level_values(0)[n] + pop.columns.get_level_values(1)[n]
              for n in range(0,len(pop.columns.get_level_values(0)))]

pop.columns = tmp_columns
pop.head()
```

```python
# 'pop' 데이터 프레임의 정보를 간략하게 본다
pop.info()
```

## 5-5 지도 시각화를 위해 지역별 고유 ID 만들기

```python
# '시도' 칼럼에 있는 고유값을 본다. -> 모든 지역을 대표하는 교우한 ID를 만든다
pop['시도'].unique()
```

```python
si_name = [None] * len(pop)

# 광역시가 아닌데 구를 가지고 있는 시와 그 행정구를 파이썬의 dict형으로 선언하여 따로 설정을 해줌
tmp_gu_dict = {'수원' : ['장안구','권선구','팔달구','영통구'],
              '성남' : ['수정구','중원구','분당구'],
              '안양':['만안구','동안구'],
              '안산':['상록구','단원구'],
              '고양':['덕양구','일산동구','일산서구'],
              '용인':['처인구','기흥구','수지구'],
              '청주':['상당구','서원구','흥덕구','청원구'],
              '천안':['동남구','서북구'],
              '전주':['완산구','덕진구'],
              '포항':['남구','북구'],
              '창원':['의창구','성산구','진해구','마산합포구','마산회원구'],
              '부천':['오정구','원미구','소사구']}
```

```python
for n in pop.index:
    # '광역시도' 칼럼 뒤에서 3자리가 '광역시','특별시','자치시'가 아닌 경우에
    if pop['광역시도'][n][-3:] not in ['광역시','특별시','자치시']:
        # '시도'칼럼에서 '고성'이고 '광역시도'칼럼에서 '강원도'일 경우
        if pop['시도'][n][:-1]=='고성' and pop['광역시도'][n] == '강원도':
            # 이름을 '고성(강원)'으로 해준다
            si_name[n] = '고성(강원)'
        # 또는 '시도'칼럼에서 '고성'이고 '광역시도'칼럼에서 '경상남도'일 경우
        elif pop['시도'][n][:-1]=='고성' and pop['광역시도'][n] == '경상남도':
            # 이름을 '고성(강원)'으로 해준다
            si_name[n] = '고성(경남)'
        # 그 외의 경우에는 그냥 둔다
        else:
            si_name[n] = pop['시도'][n][:-1]
   
    # tmp_gu_dict에 들어있는 수만큼 for문을 돌려준다
    for keys, values in tmp_gu_dict.items():
        # 만약 value가 '시도'칼럼에 있다면
        if pop['시도'][n] in values:
            # '시도 '칼럼에 있는 글자 수가 2와 같다면
            if len(pop['시도'][n]) ==2:
                # ''(공백)과 '시도'칼럼에 있는 값을 더한 값을 넣어준다
                si_name[n] = keys + '' + pop['시도'][n]
            # 또는 '시도'칼럼에 있는 '마산합포구','마산회원구'가 있으면            
            elif pop['시도'][n] in ['마산합포구','마산회원구']:
                # ''(공백)과 '시도'칼럼에 있는 값을 더한 값을 넣어준다
                si_name[n] = keys + ' ' + pop['시도'][n][2:-1]
            # 그외에는
            else:
                # ''(공백)과 '시도'칼럼에 있는 값을 더한 값을 넣어준다
                si_name[n] = keys + ' ' + pop['시도'][n][:-1]
        elif pop['광역시도'][n] == '세종특별자치시':
            si_name[n] = '세종'
        else:
            if len(pop['시도'][n]) == 2:
                si_name[n] = pop['광역시도'][n][:2] + ' ' + pop['시도'][n]
            else:
                si_name[n] = pop['광역시도'][n][:2] + ' ' + pop['시도'][n][:-1]
```

## 5-8 인구 현황에서 여성 인구 비율 확인하기

```python
def drawKorea(targetData, blockedMap, cmapname):
    gamma = 0.75
    
    whitelabelmin=20
    
    datalabel = targetData
    
    tmp_max = max([np.abs(min(blockedMap[targetData])),
                  np.abs(max(blockedMap[targetData]))])
    
    vmin, vmax = -tmp_max, tmp,max
    
    mapdata = blockedMap.pivot_table(index='y',columns='x',values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata),mapdata)
    
    plt.figure(figsize=(9,11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname,
              edgecolor = '#aaaaaa', linewidth=0.5)
    
    for idx, row in blockedMap.itterrows():
        if len(row['ID'].split())==2:
             dispname = '{}\{}'.format(row['ID'].split())[0], row(['ID'].split()[1])
        elif row['ID'][:2] == '고성':
            dispname = '고성'
        else:
            dispname = row['ID']
            
        
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 10.0, 1.1
        else:
            fontsize, linespacing = 11,1.
    
        annocolor = 'white' if row[targetData] > whitelabelmin else 'black'
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5),weight='bold'.
                fontsize=fontsize, ha='center', va='center', color=annocolor,
                linespacing=linespacing)
        
        for path in BORDER_LINES:
            ys, xs = zip(*path)
            plt.plot(xs,ys, c='black',lw=2)
            
        plt.gca().invert_yaxis()
    
        plt.axis('off')

        cb = plt.colorbar(shrink=.0,aspect=10)
        cb.set_label(datalabel)

        plt.tight_layout()
        plt.show()
```

```python
# 인구수 합계를 우리나라 지도에 적용
geo_path = './data/05. skorea_munipalities_geo_simple.json'
geo_data = json.load(open(geo_path,encoding='utf-8'))

# 앞서서 folium 모듈을 사용하여 지도를 그린 것처럼 지도를 그려준다
map = folium.Map(location=[36.2002, 127.054],zoom_start=7)
map.choropleth(geo_str = geo_data,
              data = pop_folium['인구수합계'],
              columns = [pop_folium.index, pop_folium['인구수합계']],
              fill_color = "YlGnBu",
              key_on = 'feature.id')
map

# 우리나라 인구 밀집 현상 확인
```
